```python
import pandas as pd
import numpy as np
```

```python
dfs=pd.read_excel('settlement.xlsx')
dfs.head()
```

```python
dfacc=pd.read_excel('acc_statement.xlsx')
dfacc.head()
```

```python
# what is in dfs of debit but not in dfacc of credit
setlement_debit_not_in_credit=dfs[~dfs['DR'].isin(dfacc['CREDIT'])]
setlement_debit_not_in_credit.head()
```

```python
setle_debit_notin_credit_index=setlement_debit_not_in_credit.index
setledebcre=setle_debit_notin_credit_index.to_list()
setledebcre
```

```python
rows_to_highlight = setledebcre  # Remember: row 3 = index 2

# Create an Excel writer using xlsxwriter
with pd.ExcelWriter('settlement_debit_notin_acc_credit.xlsx', engine='xlsxwriter') as writer:
    dfs.to_excel(writer, index=False, sheet_name='Sheet1')
    
    workbook  = writer.book
    worksheet = writer.sheets['Sheet1']

    # Define the format for highlighted rows
    highlight_format = workbook.add_format({'bg_color': '#FFEB9C'})  # Light yellow

    # Apply formatting to specific rows (add 1 to skip header)
    for row in rows_to_highlight:
        worksheet.set_row(row + 1, None, highlight_format)  # +1 to account for header row
```
